```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV'yi oku
df = pd.read_csv('abdomen_project/outputs/logs/cls_fold0.csv')

# Trend grafikleri
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# mAP trendi
axes[0,0].plot(df['epoch'], df['mAP'], marker='o', color='blue')
axes[0,0].set_title('mAP Trendi')
axes[0,0].set_xlabel('Epoch')
axes[0,0].set_ylabel('mAP')
axes[0,0].grid(True)

# macroF1 trendi
axes[0,1].plot(df['epoch'], df['macroF1'], marker='o', color='green')
axes[0,1].set_title('macroF1 Trendi')
axes[0,1].set_xlabel('Epoch')
axes[0,1].set_ylabel('macroF1')
axes[0,1].grid(True)

# Train Loss trendi
axes[1,0].plot(df['epoch'], df['train_loss'], marker='o', color='red')
axes[1,0].set_title('Train Loss Trendi')
axes[1,0].set_xlabel('Epoch')
axes[1,0].set_ylabel('Train Loss')
axes[1,0].set_yscale('log')
axes[1,0].grid(True)

# Learning Rate trendi
axes[1,1].plot(df['epoch'], df['lr'], marker='o', color='purple')
axes[1,1].set_title('Learning Rate Trendi')
axes[1,1].set_xlabel('Epoch')
axes[1,1].set_ylabel('LR')
axes[1,1].set_yscale('log')
axes[1,1].grid(True)

plt.tight_layout()
plt.savefig('cls_fold0_trends2.png', dpi=150)
plt.close()
print('Grafik kaydedildi: cls_fold0_trends.png')

# Sınıf bazlı final performans
classes = ['acute_cholecystitis', 'kidney_ureter_stone', 'acute_pancreatitis', 'aortic_aneurysm_dissection', 'acute_appendicitis', 'acute_diverticulitis']
final_ap = df.iloc[-1][[f'AP/{c}' for c in classes]]
final_f1 = df.iloc[-1][[f'F1/{c}' for c in classes]]

print('\n' + '='*60)
print('SON EPOCH (49) SINIF BAZLI PERFORMANS:')
print('='*60)
for c, ap, f1 in zip(classes, final_ap, final_f1):
    print(f'{c:25} | AP: {ap:.4f} | F1: {f1:.4f}')
```
