```python
import sys
from pathlib import Path
import pydicom

# Veri seti kontrol
TRAIN_DIR = Path("/Users/ramazanpolat/Desktop/datasets/abdomen/Eğitim Verisi")
TEST_DIR = Path("/Users/ramazanpolat/Desktop/datasets/abdomen/Yarışma Veri Seti")

print("=" * 80)
print("TotalSegmentator UYGUNLUK ANALİZİ")
print("=" * 80)

# 1. Veri Seti Yapısı
print("\n1️⃣ VERİ SETİ YAPISI:")
print(f"   Eğitim Verisi: {TRAIN_DIR.exists()} ({TRAIN_DIR})")
print(f"   Test Verisi: {TEST_DIR.exists()} ({TEST_DIR})")

# 2. DICOM Dosya Sayısı
if TRAIN_DIR.exists():
    dcm_files_train = list(TRAIN_DIR.glob("**/*.dcm"))
    print(f"\n   📊 Eğitim Verisi DICOM sayısı: {len(dcm_files_train):,}")
    
if TEST_DIR.exists():
    dcm_files_test = list(TEST_DIR.glob("**/*.dcm"))
    print(f"   📊 Test Verisi DICOM sayısı: {len(dcm_files_test):,}")

# 3. DICOM Dosya Örneği İncelemesi
print("\n2️⃣ DICOM ÖRNEĞİ İNCELEMESİ:")
if dcm_files_train:
    sample_dcm = dcm_files_train[0]
    print(f"   Örnek dosya: {sample_dcm.name}")
    
    try:
        ds = pydicom.dcmread(sample_dcm)
        print(f"   ✅ DICOM okunabilir")
        print(f"   • Modality: {ds.Modality if 'Modality' in ds else 'N/A'}")
        print(f"   • Shape: {ds.pixel_array.shape}")
        print(f"   • Pixel Spacing: {ds.PixelSpacing if 'PixelSpacing' in ds else 'N/A'}")
        print(f"   • Slice Thickness: {ds.SliceThickness if 'SliceThickness' in ds else 'N/A'}")
        
        # TotalSegmentator gereksinimlerini kontrol et
        print(f"\n3️⃣ TotalSegmentator GEREKSİNİMLERİ:")
        modality = str(ds.Modality) if 'Modality' in ds else "Unknown"
        
        if modality == 'CT':
            print(f"   ✅ Modality: CT (Uygun)")
        else:
            print(f"   ❌ Modality: {modality} (Uygun değil - CT gerekli)")
        
        # Voxel spacing check
        if 'PixelSpacing' in ds and 'SliceThickness' in ds:
            px_spacing = float(ds.PixelSpacing[0])
            slice_thick = float(ds.SliceThickness)
            print(f"   ✅ Voxel Spacing: ({px_spacing:.2f}, {px_spacing:.2f}, {slice_thick:.2f}) mm")
        
        # Image dimensions
        shape = ds.pixel_array.shape
        if len(shape) == 2:
            print(f"   ✅ Image Size: {shape[0]} x {shape[1]} pixels (2D slice)")
        else:
            print(f"   Image Size: {shape}")
            
    except Exception as e:
        print(f"   ❌ Hata: {e}")

print(f"\n4️⃣ TotalSegmentator UYGUNLUK DURUMU:")
print("""
   ✅ İDEAL SENARYO:
   • DICOM formatı ✓
   • CT Modality ✓
   • Standard voxel spacing ✓
   
   ⚠️ NOT: TotalSegmentator tam abdomen CT serileri (3D volume) bekler.
      Tek kesit (2D) veya düşük çözünürlüklü görsellerde çalışmaz.
   
   🔍 KONTROL NOKTASI:
      • Her vaka kaç kesit içeriyor?
      • Kesitler birbirine bağlı mı (seri)?
      • Pixel spacing standardmı?
""")

# 5. Vaka başına kesit sayısı
print(f"\n5️⃣ VAKA BAŞINA KESİT SAYISI:")
if TRAIN_DIR.exists():
    case_slices = {}
    for dcm_file in dcm_files_train[:100]:  # İlk 100'ü kontrol et
        case_num = dcm_file.parent.name
        case_slices[case_num] = case_slices.get(case_num, 0) + 1
    
    if case_slices:
        avg_slices = sum(case_slices.values()) / len(case_slices)
        min_slices = min(case_slices.values())
        max_slices = max(case_slices.values())
        print(f"   • Ortalama: {avg_slices:.0f} kesit/vaka")
        print(f"   • Min: {min_slices} kesit/vaka")
        print(f"   • Max: {max_slices} kesit/vaka")
        
        if avg_slices > 50:
            print(f"   ✅ Yeterli kesit sayısı (TotalSegmentator için ideal)")
        elif avg_slices > 20:
            print(f"   ⚠️ Orta düzey kesit sayısı")
        else:
            print(f"   ❌ Yetersiz kesit sayısı")
```

```python
from pathlib import Path
import pydicom

TRAIN_DIR = Path("/Users/ramazanpolat/Desktop/datasets/abdomen/Eğitim Verisi")
dcm_files_train = list(TRAIN_DIR.glob("**/*.dcm"))

print("=" * 80)
print("TotalSegmentator UYGUNLUK ANALİZİ")
print("=" * 80)

print(f"\n1. VERİ SETİ YAPISI:")
print(f"   Eğitim Verisi DICOM sayısı: {len(dcm_files_train):,}")

if dcm_files_train:
    sample_dcm = dcm_files_train[0]
    print(f"\n2. DICOM ÖRNEĞİ: {sample_dcm.name}")
    
    try:
        ds = pydicom.dcmread(sample_dcm)
        print(f"   ✅ DICOM okunabilir")
        modality = str(ds.Modality) if 'Modality' in ds else 'N/A'
        print(f"   Modality: {modality}")
        print(f"   Shape: {ds.pixel_array.shape}")
        if 'PixelSpacing' in ds:
            print(f"   Pixel Spacing: {ds.PixelSpacing}")
        if 'SliceThickness' in ds:
            print(f"   Slice Thickness: {ds.SliceThickness}")
    except Exception as e:
        print(f"   ❌ Hata: {e}")

# Vaka başına kesit sayısı
case_slices = {}
for dcm_file in dcm_files_train[:100]:
    case_num = dcm_file.parent.name
    case_slices[case_num] = case_slices.get(case_num, 0) + 1

if case_slices:
    avg_slices = sum(case_slices.values()) / len(case_slices)
    min_slices = min(case_slices.values())
    max_slices = max(case_slices.values())
    print(f"\n3. VAKA BAŞINA KESİT SAYISI:")
    print(f"   Ortalama: {avg_slices:.0f} kesit/vaka")
    print(f"   Min: {min_slices}, Max: {max_slices}")
```
